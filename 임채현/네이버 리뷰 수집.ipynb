```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time


driver_path = 'C:/Users/임채현/[숙탯]겨울/chromedriver.exe'

# ChromeOptions 설정
chrome_options = webdriver.ChromeOptions()


# 다른 옵션들을 추가하려면 여기에 추가
#service = Service(executable_path=driver_path)

# 크롬 드라이버 초기화
driver = webdriver.Chrome( options=chrome_options)
driver.get('https://map.naver.com/v5/search')

# 팝업 창 제거
#driver.find_element_by_css_selector("button#intro_popup_close").click()


time.sleep(5)
search_input = driver.find_element(By.CSS_SELECTOR, 'div.input_box input.input_search')
search_input.click()
#식당 이름 도로명주소 도로명까지만 입력해서 search (주소 다 치면 안 나오더라구요..)
search_input.send_keys("맥주창고 제주특별자치도 제주시 도남로")

time.sleep(3)

# 검색버튼 누르기
search_input.send_keys(Keys.ENTER)
time.sleep(5)


frame_xpath = '//*[@id="entryIframe"]'
frame = WebDriverWait(driver, 10).until(
    EC.frame_to_be_available_and_switch_to_it((By.XPATH, frame_xpath))
)



# WebDriverWait를 사용하여 요소가 나타날 때까지 대기
review_link_xpath = '//*[@id="app-root"]/div/div/div/div[4]/div/div/div/div/a[2]'
//*[@id="app-root"]/div/div/div/div[4]/div/div/div/div/a[2]

# 요소가 나타날 때까지 대기하고 클릭 가능할 때까지 기다림
review_link = WebDriverWait(driver, 20).until(
    EC.element_to_be_clickable((By.XPATH, review_link_xpath))
)

# 링크 클릭
review_link.click()


while True:
    try:
        # "더보기" 버튼을 찾고 클릭
        more_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="app-root"]/div/div/div/div[6]/div[3]/div[3]/div[2]/div/a'))
        )
        more_button.click()

        # 리뷰 로드 대기 (10초 동안 기다림)
        time.sleep(10)

        # 현재까지 수집한 리뷰 개수 확인
        current_review_count = len(driver.find_elements(By.XPATH, '//ul[@class="eCPGL"]/li[@class="YeINN"]'))
        
        # 수집한 리뷰가 100개 이상이라면 반복 종료
        if current_review_count >= 100:
            break

    except Exception as e:
        # 더이상 "더보기" 버튼이 없거나 예외가 발생하면 반복 종료
        print("No more reviews or an error occurred.")
        break

# 모든 리뷰 요소들을 찾기
review_elements = WebDriverWait(driver, 20).until(
    EC.presence_of_all_elements_located((By.XPATH, '//ul[@class="eCPGL"]/li[@class="YeINN"]'))
)

# 각 리뷰의 텍스트 수집하기 (최대 100개)
for idx, review_element in enumerate(review_elements[:100], 1):
    review_text = review_element.find_element(By.XPATH, './/div[@class="ZZ4OK IwhtZ"]/a/span[@class="zPfVt"]').text
    print(f'Review {idx}: {review_text}')
```
